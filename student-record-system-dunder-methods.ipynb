```python
class Student():

    def __init__(self, name, marks, branch):
          self.name = name
          self.marks = marks
          self.branch = branch


    def __str__(self):
        return f"{self.name} | {self.marks} | {self.branch}"
    
    def __len__(self):
        return len(self.name)
    
student = Student("Jaydeep", 95, "Extc")

print(student)
print(len(student))
```
